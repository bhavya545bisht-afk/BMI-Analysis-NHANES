```python
import matplotlib.pyplot as plt

# Representing minimum as well as maximum body weight measurements of adult males as well as females with the help of histogram

plt.subplot(2,1,1)
plt.hist(female_weights, bins=20)
plt.title("Female Weights")

plt.subplot(2,1,2)
plt.hist(male_weights, bins=20)
plt.title("Male Weights")

# Make same x limits
xmin = min(female_weights.min(), male_weights.min())
xmax = max(female_weights.max(), male_weights.max())

plt.subplot(2,1,1)
plt.xlim(xmin, xmax)

plt.subplot(2,1,2)
plt.xlim(xmin, xmax)

plt.tight_layout()
plt.show()
```

```python
plt.boxplot([female_weights, male_weights],
            tick_labels=["Female", "Male"])

# Representing the body weight measurements of adult  males as well as females using boxplot

plt.title("Weight Comparison")
plt.show()
```

```python
def describe(data):
    return {
        "Mean": np.mean(data),
        "Median": np.median(data),
        "Std": np.std(data),
        "Min": np.min(data),
        "Max": np.max(data),
        "Skew approx": np.mean((data - np.mean(data))**3) / (np.std(data)**3)
    }
# Calculating statistical data of body weight  measurements of adult  males as well as females

print("Female stats:", describe(female_weights))
print("Male stats:", describe(male_weights))
```

```python
import pandas as pd

female = pd.read_csv("nhanes_adult_female_bmx_2020.csv", comment="#")
```

```python
female["Height_m"] = female["BMXHT"] / 100
female["BMI"] = female["BMXWT"] / (female["Height_m"] ** 2)
#Adding BMI column in Adult Females dataset(incluging their all body measurements ) 
```

```python
cols = ["BMXWT","BMXHT","BMXARML","BMXLEG",
        "BMXARMC","BMXHIP","BMXWAIST","BMI"]

zfemale = (female[cols] - female[cols].mean()) / female[cols].std()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Representing a body measurements of adult females using a scattered plot such as body weight,body height,body upperarmlength,body upperleglength etc.
sns.pairplot(zfemale)
plt.show()
```

```python
print(zfemale.corr(method="pearson"))                    #Correlation of pearson from adult females dataset (including there all body measurements)

```

```python
print(zfemale.corr(method="spearman"))                      #Correlation of spearman from adult females dataset (including there all body measurements)
```

```python
female["WHtR"] = female["BMXWAIST"] / female["BMXHT"]                 # Waist-to-Height = Waist circumference / Height              (Adult Females)
female["WHR"] = female["BMXWAIST"] / female["BMXHIP"]                 # Waist-to-Hip = Waist circumference / Hip circumference      (Adult Females)
                                                                      
                                                                      
```

```python
male = pd.read_csv("nhanes_adult_male_bmx_2020.csv", comment="#")

male["WHtR"] = male["BMXWAIST"] / male["BMXHT"]                       # Waist-to-Height = Waist circumference / Height             (Adult Males)
male["WHR"] = male["BMXWAIST"] / male["BMXHIP"]                       # Waist-to-Hip = Waist circumference / Hip circumference     (Adult Males)
```

```python
import matplotlib.pyplot as plt

# Comparing the ratios of waist-to-height and waist-to-hip of adult males and females using boxplot 

plt.boxplot([
    female["WHtR"],
    male["WHtR"],
    female["WHR"],
    male["WHR"]
],
tick_labels=["Female WHtR", "Male WHtR", "Female WHR", "Male WHR"])

plt.title("Ratio Comparison")
plt.show()
```

```python
                                                    Advantages of Body Mass Index (BMI)

1.Very easy to compute

* BMI only needs weight and height, which are available for everyone in our dataset.

* Thats why BMI is practical for large surveys like NHANES with thousands of participants.

2.Quick population-level comparison

From our stats:

* Male mean BMI is higher than female mean BMI

* This allows easy comparison between groups (male vs female).

* Public health researchers can quickly identify obesity trends.

3.Standardized and widely used

* BMI categories (underweight, normal, overweight, obese) are globally recognized.

* Results can be compared with other studies and countries.

4.Correlates reasonably well with health risk (on average)

In our data, higher BMI generally corresponds to:

* Higher waist circumference

* Higher hip circumference

This suggests BMI is a rough proxy for body fat at the population level.

    

                                             Disadvantages of Body Mass Index (BMI)

1.Does NOT distinguish fat from muscle

Males in our dataset tend to:

* Have higher weight

* Larger arm and leg measurements

* A muscular male can have a high BMI but low body fat, leading to misclassification as overweight.

2.Ignores fat distribution

* BMI treats all weight the same.

But our data shows:

* Waist circumference varies a lot even at similar BMI values.

* Central fat (abdominal fat) is more dangerous than fat elsewhere, and BMI cannot detect this.

3.Sex differences are not handled well

Females typically:

* Have higher body fat percentage at the same BMI

* Store fat differently (hips vs waist)

* BMI uses the same cutoffs for males and females, which can be misleading.

4.Height-related distortion

* Very tall or very short individuals can have BMI values that do not reflect true body composition.

* This matters in our dataset where height varies substantially.
    
```

```python
cols = ["BMXWT","BMXHT","BMXARML","BMXLEG",
        "BMXARMC","BMXHIP","BMXWAIST","BMI"]

zfemale = (female[cols] - female[cols].mean()) / female[cols].std()
#Ensuring that our females should have a bmi column
```
